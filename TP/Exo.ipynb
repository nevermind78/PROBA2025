```python
Alpha =''.join((chr(i) for  i in range(65,75)))
Alpha
```

```python
EnsL = {w+'L' for w in Alpha}
EnsR = {w+'R' for w in Alpha}
```

```python
ENSEMBLE = EnsL.union(EnsR)
print(ENSEMBLE)
```

```python
from itertools import combinations
Combinaisons = list(combinations(ENSEMBLE, 4))
print(len(Combinaisons))
```

```python
from math import factorial
factorial(20)//(factorial(4)*factorial(16))
```

```python
print(Combinaisons)
```

```python
def nbr_paires_chaussettes(t):
    d= set(w[0] for w in t)
    return 4-len(d)
```

```python
nbr_paires_chaussettes(('CL','AR','BL','DR'))
```

```python
len(list(filter(lambda x: nbr_paires_chaussettes(x)==1, Combinaisons)))
```
